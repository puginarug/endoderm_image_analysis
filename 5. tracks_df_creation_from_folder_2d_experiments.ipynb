# Imports

```python
import numpy as np
import matplotlib.pyplot as plt
import btrack
import pandas as pd
import seaborn as sns
import os
```

# Define file locations and read data

```python
WORKING_DIR = r'D:\David\endoderm_migration\raw_data\2025_01_22_2DFACS_FOXA2positive_lateday2' # example: 'd:\\David\\endoderm_migration\\2024_09_09_FoxA2_tdTom_c211to300_BmpAct_day2To3'
TRACKS_FOLDER = WORKING_DIR+'\\btrack_tracks'

# check available track files
track_files = [f for f in os.listdir(TRACKS_FOLDER) if f.endswith('.h5')]
print("Available track files:", track_files)
```

## pixel size

```python
# pixel size in microns to convert the pixel coordinates to microns
pixel_size = 0.32 # microns per pixel
```

```python
# open one of the track files
with btrack.io.HDF5FileHandler(
    TRACKS_FOLDER+'\\'+track_files[5], 'r', obj_type='obj_type_1'
) as reader:
    tracks = reader.tracks
```

# Tracks

```python
# load tracks generated by btrack
tracks_dict = {}
for filename in os.listdir(TRACKS_FOLDER):
    if filename.endswith(".h5"):
      with btrack.io.HDF5FileHandler(
         TRACKS_FOLDER+'\\'+filename, 'r', obj_type='obj_type_1'
      ) as reader:
        print(f"Loaded {filename}")
        tracks_dict[filename] = reader.tracks
```

```python
reader.tracks
```

```python
# get the x and y coordinates of the tracks
track_coordinates = {}

for filename, tracks in tracks_dict.items():
    print(filename)
    track_coordinates[filename] = []
    for track in tracks:
        x = track['x']
        y = track['y']
        track_coordinates[filename].append(np.column_stack((x, y)))
```

## Creating df

```python
dt = 5 # time interval between frames in minutes
```

```python
# create a DataFrame with all the coordinates and track IDs
tracks_df = pd.DataFrame()

for filename, tracks in tracks_dict.items():
    for track in tracks:
        # Extract the data
        ID = track['ID']
        time_points = track['t']
        x_coords = track['x']
        y_coords = track['y']
        major_axis_length = track['major_axis_length']
        minor_axis_length = track['minor_axis_length']

        # Create a DataFrame
        data = {
            'file': [filename] * len(time_points),
            'ID': [ID] * len(time_points),
            't': time_points,  # convert to actual time using dt
            'x_pixels': x_coords,
            'y_pixels': y_coords,
            'major_axis_length': major_axis_length,
            'minor_axis_length': minor_axis_length
        }

        tracks_df = pd.concat([tracks_df, pd.DataFrame(data)])

# calculate the normalized time for each track
tracks_df['t'] = tracks_df['t'] * dt # convert to actual time using dt
tracks_df['normalized_time'] = tracks_df.groupby(['ID','file'])['t'].transform(lambda x: x - x.min())

tracks_df['step'] = tracks_df['normalized_time']// dt

tracks_df.reset_index(drop=True, inplace=True)

# combine file and ID to create a unique track ID
tracks_df['track_id'] = tracks_df['file'] + '_' + tracks_df['ID'].astype(str)

# convert units to microns
tracks_df['x_microns'] = tracks_df['x_pixels'] * pixel_size
tracks_df['y_microns'] = tracks_df['y_pixels'] * pixel_size
tracks_df['major_axis_length_microns'] = tracks_df['major_axis_length'] * pixel_size
tracks_df['minor_axis_length_microns'] = tracks_df['minor_axis_length'] * pixel_size

# reorganize the DataFrame
tracks_df = tracks_df[['file', 'ID', 'track_id', 't', 'normalized_time', 'step', 'x_pixels', 'y_pixels', 'x_microns', 'y_microns', 'major_axis_length_microns', 'minor_axis_length_microns']]

tracks_df.head()
```

```python
# save the DataFrame to a csv file
save_to = r'D:\\David\\THESIS\\tracks_dataframes'
tracks_df.to_csv(save_to+f'\\centers_segmentation_tracks_df_{WORKING_DIR.split("\\")[-1]}.csv', index=False)

print('Tracks saved to csv file: ', save_to+f'\\tracks_df_{WORKING_DIR.split("\\")[-1]}.csv')
```

# Cell Counts

```python
# create a df with cell counts per timepoint (rows: timepoints, columns: file, values: cell counts)
cell_counts = tracks_df.groupby(['file', 't']).size().unstack().fillna(0)

# plot the cell counts per timepoint
plt.figure(figsize=(10, 5))
sns.heatmap(cell_counts, cmap='viridis', cbar_kws={'label': 'Cell Count'})
plt.xlabel('Timepoint')
plt.ylabel('File')
plt.title('Cell Counts Per Timepoint')
```
